# Data Analysis and Visualization

```python
# to check the % of cancelled booking  using value_count(normalize = True) function 
#normalize if used to get value in percent

cancelled_perc = df['is_canceled'].value_counts(normalize = True )
cancelled_perc
```

```python
#Show the reservation status count using plt

print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title("Reservation Status Count")
plt.bar(["Not Cancelled","Cancelled"],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.7)
plt.show()
```

The above chart shows the comparison of Cancelled and Not Cancelled Reservation

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x = 'hotel',hue = 'is_canceled',data = df,palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(loc='upper right', bbox_to_anchor=(0, 5))
plt.title("Reservation status in different hotels",size = 20)
plt.xlabel("hotel")
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

The above chart shows the cancellation count in both the hotels

```python
#to check the % of cancellation in each hotel

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#to find the change in rate per reservation in both the hotels

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (30,10))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = "Resort Hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label = "Resort Hotel")
```

Average daily rate rise and fall in each hotel

```python
df['month'] = df['reservation_status_date'].dt.month
df
```

# Analysis with month data for reservation status and rates.

```python
plt.figure(figsize = (16,8))
ax2 = sns.countplot(x = 'month', hue = 'is_canceled',data = df)
ax2.legend(bbox_to_anchor=(1,1))
plt.title("Reservation Status per Month",size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

The graph above compares cancelled and uncancelled bookings on a month-by-month basis.

```python
data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
ax=sns.barplot(x = 'month',y= 'adr' ,data=data)
plt.title("Monthwise Average Daily Rate Change",fontsize = 20)
```

Here the month on month daily rate changes are compared

```python
cancelled_data = df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 countries with reservation cancelled",size = 20)
plt.pie(top_10_country,autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

Following chart shows the top 10 countries with most cancelled reservations
